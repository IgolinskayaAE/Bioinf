В NCBI(https://www.ncbi.nlm.nih.gov/datasets/genome/) нашла геномы двух организмов - человека и дрозофилы, хочу найти гены с определенными мотивами.

Скачала их геномы по названиями humanrho.fasta и drorho.fasta соответсвенно в папку ncbi_dataset в домашней директории с помлщью команды datasets download genome accession <id генома вашего любимого организма> --include protein

На сайте https://www.ebi.ac.uk/interpro/entry/pfam/#table нашла HMM для белкового домена - PF00001 (7tm_1, родопсин-подобный белок). Этот домен характерен для белков семейства G-белковых рецепторов

с помощью хммер ищем белки, содержащие этот домен в геномах обоих организмов

```python
#команда для человека
hmmsearch --tblout human_results.tbl PF00001.hmm humanrho.fasta
```

```python
#и дрозофилы
hmmsearch --tblout drosophila_results.tbl PF00001.hmm drorho.fasta
```
